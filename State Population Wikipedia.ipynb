```python
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import random
from mpl_toolkits.mplot3d import Axes3D
from bs4 import BeautifulSoup
import urllib
%matplotlib inline
```

```python
#read page
page = urllib.urlopen('https://en.wikipedia.org/wiki/List_of_U.S._states_and_territories_by_population').read()
soup = BeautifulSoup(page, "lxml")
print soup.prettify()[10000:10000000]
```

```python
#getting all the 'a' tags
states_names = soup.table.find_all('a')
```

```python
#iterate through the state list
states = []
for i in range(55):
    #each time, add the state name to list
    states.append(states_names[i+2]['title'])
```

```python
#adding the last U.S. territory 
states.append(states_names[59]['title'])
```

```python
#getting the total population from table
table = soup.find( "table", {"class":"wikitable sortable"} )
numbers = soup.find_all("td")
```

```python
#extracting population number for each state.
#pop number appears every 9 entries
population = []
for i in range(3,1274,9):
    num = numbers[i].get_text()
    population.append(num)
```

```python
#making each entry into a string
tot_population = []
for i in range(56): 
    pop = str(population[i])
    tot_population.append(pop)
print len(tot_population)
```

```python
#creating dataframe
state_population=pd.DataFrame(
    {'state_name':states,
    'estimated_population': tot_population,
    })
```
